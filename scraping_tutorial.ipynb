```python
from selenium import webdriver
driver = webdriver.Chrome(r"C:\Users\anmol\Desktop\chromedriver.exe")
```

```python
driver.get("https://www.kinopoisk.ru/lists/series-top250/?tab=all")
```

как интерактировать со страницей, самое полезное: 
* можно нажимать на кнопочки: найти, потом клинкуть
* можно вводить логин/пароль: найти, потом send_keys

найти элементы: (список функций)[https://selenium-python.readthedocs.io/locating-elements.html], также можно делать это с помощью css или класса

```python
driver.find_element_by_css_selector(".paginator__page-relative").click()
```

давайте точно также соберем наши данные, только уже со всех страниц

```python
driver.find_element_by_css_selector(".paginator__page-relative+ .paginator__page-number").click() #вернемся на первую страницу
```

```python
# find_elements_by_xpath returns an array of selenium objects.
titles_element = driver.find_elements_by_css_selector(".selection-film-item-meta__name")
# use list comprehension to get the actual titles and not the selenium objects.
titles = [x.text for x in titles_element]
# print out all the titles.
print('titles:')
print(titles, '\n')
```

```python
# find_elements_by_xpath returns an array of selenium objects.
rating_element = driver.find_elements_by_css_selector(".rating__value_positive")
# use list comprehension to get the actual titles and not the selenium objects.
rating = [x.text for x in rating_element]
print('ratings:')
print(rating, '\n')
```

```python
driver.find_element_by_css_selector(".paginator__page-number_is-active+ .paginator__page-number").click()
```

```python
# find_elements_by_xpath returns an array of selenium objects.
titles_element1 = driver.find_elements_by_css_selector(".selection-film-item-meta__name")
# use list comprehension to get the actual titles and not the selenium objects.
titles1 = [x.text for x in titles_element1]
# print out all the titles.
print('titles:')
print(titles1, '\n')
```

```python
# find_elements_by_xpath returns an array of selenium objects.
rating_element1 = driver.find_elements_by_css_selector(".rating__value_positive")
# use list comprehension to get the actual titles and not the selenium objects.
rating1 = [x.text for x in rating_element1]
print('ratings:')
print(rating1, '\n')
```

```python
rating_fin = rating + rating1
#print(rating_fin)
```

```python
title_fin = titles+titles1
#print(title_fin)
```

```python
import pandas as pd
wow = pd.DataFrame({'title': title_fin, 'rating': rating_fin})
wow
```

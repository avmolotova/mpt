# Функции!

A function is a named sequence of statements. It helps the programmer better think about the solution by organizing pieces of the solution into logical chunks that can be reused 


## Общий синтаксис

```
def name(formal parameters ):
    """docstrings"""
    stat = statements
    return (stat)

name(actual parameters) # чтобы вывести функцию
```

**имя** - любое

**формальные параметры** - что нужно функции, чтобы она работала (*= inputs of a function*). Может содержать один, несколько параметров(указываются через запятую) или быть пустым . Мы можем манипулировать параметрами в теле функции (и только там!)

**действительные параметры** - значения, которые мы хотим прогнать через нашу функцию

**docstring** = строка документации для функции

**return** - immediately ends execution of the code block inside a function. Используется, чтобы вывести значение (value). Это опционально (иногда функции используются, чтобы просто осуществить действие), в зависимости от цели функции

````
def square_fun(x):
    y = x * x
    return y

square_fun(5)
````

## Accumulative functions

Ничего нового, просто закидываем аккумуляторы в функцию


````
# функция для нахождения максимального значения

def acc_fun(seq):
    c = 0
    for i in seq:
        if i > c:
          c = i
    return c

print(acc_fun([1, 2, 13, 7]))
````
